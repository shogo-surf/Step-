```python
import pandas as pd
df = pd.read_csv("meguro_raw_data.csv")
```

```python
df.head()
```

```python
#数値データへの変換
#「家賃」「管理費」「敷金」「礼金」「面積」「築年数」
import re
def get_number(value):
    n = re.findall(r"[0-9.]+", value)
    
    if len(n)!=0:
        return float(n[0])
    else:
        return 0

df["家賃"] = df["家賃"].apply(get_number)
df["管理費"] = df["管理費"].apply(get_number)
#単位を万円に統一
df["管理費"] = df["管理費"] / 10000
df["敷金"] = df["敷金"].apply(get_number)
df["礼金"] = df["礼金"].apply(get_number)
df["面積"] = df["面積"].apply(get_number)
df["築年数"] = df["築年数"].apply(get_number)

```

```python
df.head()
```

```python
#アクセスから「沿線」「駅」「徒歩」を抽出(徒歩は数値データに変換)
# 沿線を抽出
df["沿線"] = df["アクセス"].apply(lambda x: x.split("/")[0])

# アクセスから駅を抽出
df["駅"] = df["アクセス"].apply(lambda x: x.split(" ")[0].split("/")[1])

# アクセスから徒歩を抽出
df["徒歩"] = df["アクセス"].apply(lambda x: int(re.findall(r"[0-9]+", x.split(" ")[1])[0]))
```

```python
df.head()
```

```python
#重複の排除

# "URL", "アクセス"をキーに重複データを削除
df.drop_duplicates(subset=["URL", "アクセス"], inplace=True)
```

```python
#その他もキーに重複を削除
df.drop_duplicates(subset=["名称", "カテゴリー", "アドレス", "築年数", "構造", "階数", "家賃", "管理費", "敷金", "礼金", "間取り", "面積", "沿線", "駅", "徒歩"], inplace=True)
```

```python
df.to_csv("meguro_raw_data_整形.csv")
```
